```python
import kagglehub
import pandas as pd

# Download latest version of the dataset
path = kagglehub.dataset_download("yakhyojon/tiktok")
print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

# 1. Find the exact CSV file inside the folder Kaggle downloaded
dataset_files = os.listdir(path)
csv_filename = [file for file in dataset_files if file.endswith('.csv')][0]
full_file_path = f"{path}/{csv_filename}"

# 2. Load the data into a Pandas DataFrame
df = pd.read_csv(full_file_path)
print(f"✅ Successfully loaded {len(df)} TikTok videos into the engine.\n")

# 3. Calculate the "Stickiness Ratio" (Shares + Downloads per View)
# We add a small number (0.0001) to prevent dividing by zero
df['stickiness_ratio'] = (df['video_share_count'] + df['video_download_count']) / (df['video_view_count'] + 0.0001)

# 4. Identify the "Hidden Gems"
# (Unverified creators who have high stickiness ratios)
hidden_gems = df[df['verified_status'] == 'not verified'].sort_values(by='stickiness_ratio', ascending=False)

# Display the top 5 highest-potential videos
print("🔥 TOP 5 'RISING STAR' VIDEOS IDENTIFIED:")
display(hidden_gems[['video_id', 'video_view_count', 'video_share_count', 'stickiness_ratio']].head())
```
